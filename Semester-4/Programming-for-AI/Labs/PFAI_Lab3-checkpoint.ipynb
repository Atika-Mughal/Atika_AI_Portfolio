### Task 1

```python
data_dict = {}
std_marks = []
no_students = int(input("Enter the number of students:"))
for i in range(no_students):
    name = input(f"Enter the name of student {i+1}: ")
    attendance = int(input(f"Enter the attendance % {i+1}:"))
    marks = int(input(f"Enter the marks: {i+1}"))

    grade = ""
    if marks >= 80:
        grade = 'A'
    elif marks >= 70:
        grade = 'B'
    elif marks >= 60:
        grade = 'C'
    else:
        grade = 'F'
        
    data_dict[name] = {"attendance": attendance, "marks": marks, "grade": grade}
    print(data_dict)

    std_marks.append(marks)

for key, value in data_dict.items():
    if value ["attendance"] >= 75:
        print(key)
fail_count = 0
total = 0
for marks in std_marks:
    total += marks
    if marks < 50:
        fail_count += 1
print("Failures:", fail_count)
print("Class Average:", total / len(std_marks))
```

### Task 2

```python
expenses = []
total = 0
average = 0
expense = input("Enter the daily expense")
while expense != "done":
    expenses.append(int(expense))
    expense = input("Enter expense:")

for i in expenses:
    total += i
average = total / len(expense)
print("\nTotal Expense:", total)
print("Average:", average)

lst = [x for x in expenses if x > average]
print("\nMax num:", lst)
lst = [x for x in expenses if x < average]
print("Min num:", lst)
tuple = total, average
print("\nTuple:", tuple)
```

### Task 3

```python
feedbacks = [
    "The course was excellent and very good",
    "Good content but the pace was average",
    "The lectures were poor and bad",
    "Excellent explanation and good examples",
    "Average course but good instructor"
]
keywords = ['excellent', 'good', 'average', 'bad', 'poor']
words = {w: 0 for w in keywords}
positive = 0
neutral = 0
negative = 0

for fd in feedbacks:
    fd = fd.lower().strip()
    
    for w in words:
        if w in words:
            words[w] += 1
        
    if 'excellent' in fd or 'good' in fd:
        category = "Positive"
        positive += 1

    elif 'average' in fd:
        category = "Neutral"
        neutral += 1
        
    elif 'bad' in fd or 'poor' in fd:
        category = "Negative"
        negative += 1
        
    print("\nFeedback:", fd)
    print("Category:", category)
    
print("\nSummary:")
print("Positive Words:", positive)
print("Negative Words:", negative)
print("Neutral Words:", neutral)
```

### Task 4

```python
ride_types = {"mini": 25, "sedan": 40, "suv": 60}
peak_hours = (18, 19, 20, 21)
rides = []
distance = 0
hour = 0
total_fare = 0
total = 0
ride_details = {}
while True:
    distance = int(input("Enter the distance:"))
    if distance == 0:
        break
    hour = int(input("Enter the hour 0-23:"))
    type = input("Enter the Ride type:").lower()

    if type not in ride_types:
        print("Invalid")
        continue
    rate = ride_types[type]
    fare = distance * rate
    if hour in peak_hours:
        fare = fare + (fare * 0.20)

    ride_detail = {
        "distance": distance,
        "hour": hour,
        "ride_type": type,
        "fare": fare
    }
    rides.append(ride_detail)
    total_fare += fare
print("\nDetails")
for ride in rides:
    print(ride)
print("Total fare earned:", total_fare)
```
