### Task 1

```python
def safe_division(a, b):
    try:
        return a/b
    except ZeroDivisionError:
        print("It cannot divide by zero.")
    except TypeError:
        print("Invalid input! Enter a number.")
    finally:
        print("Operation Completed.")

a = float(input("Enter value for a:"))
b = float(input("Enter value for b:"))

result = safe_division(a, b)
print("Result:", result)
```

### Task 2

```python
def validate_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative.")
    elif age > 120:
        raise ValueError("Age cannot be greater than 120.")
    else:
        print("Valid Age.")

try:
    age = int(input("Enter age: "))
    validate_age(age)
except ValueError as e:
    print("Error:", e)
```

### Task 3

```python
class InsufficientBalanceError(Exception):
    pass
    
def withdraw(amount):
    balance = 5000
    if amount > balance:
        raise InsufficientBalanceError("Insufficient Balance.")
    elif amount <= 0:
        raise ValueError("Amount cannot be negative.")
    else:
        remaining_balance = balance - amount
        print("Remaining Balance:", remaining_balance)

try:
    amount = float(input("Enter the amount:"))
    withdraw(amount)

except InsufficientBalanceError as e:
    print("Error:", e)
    
except ValueError as e:
    print("Error:", e)
```

### Task 4

```python
class OutOfStockError(Exception):
    pass
    
products = {
    "laptop": 80000,
    "mouse": 1500,
    "keyboard": 3000
}

def add_to_cart(product_name, quantity):
    if product_name not in products:
        raise KeyError("Product is not in dictionary.")
    elif quantity <= 0:
        raise ValueError("Quantity cannot be negative.")
    elif quantity > 5:
        raise OutOfStockError("Choose quantity less than 5.")
    else:
        print("Product added to cart.")

try:
    product_name = input("Enter the product name:")
    quantity = int(input("Enter the quantity"))
    add_to_cart(product_name, quantity)
    
except KeyError as e:
    print("Error:", e)

except ValueError as e:
    print("Error:", e)

except OutOfStockError as e:
    print("Error:", e)
```
