### Encapsulation

```python
class ATM: 
    def __init__(self, pin):
        self.__pin = pin

    # Getter
    @property
    def pin(self):
        return self.__pin

    # Setter
    @pin.setter
    def pin(self, input):
        if input == 1234:
            self.__pin = input
        else:
            print("Wrong Input")
atm = ATM(1234)
print(atm.pin)
atm.pin = 4444
```

### Inheritance

```python
# Simple Inheritance
class Birds:
    def speak(self):
        print("Birds Voice.")

class Parrot(Birds):
    def voice(self):
        print("Parrot is here!")

b = Birds()
b.speak()
p = Parrot()
p.speak()
p.voice()
```

```python
#1. Single Inheritance
class Teacher:
    def introduction(self):
        print("This is a teacher method.")

class Student(Teacher):
    def introduction(self):
        print("I am a Student of AI.")
    def study(self):
        print("Student Method.")

t = Teacher()
t.introduction()
s = Student()
s.introduction()
s.study()
```

```python
#2. Multilevel Inheritance
class VC:
    def intro(self):
        print("Vice Chairman!")

class Teacher(VC):
    def show_teacher(self):
        print("Here is a Teacher!")

class Student(Teacher):
    def show_student(self):
        print("Student method.")

v = VC()
v.intro()
t = Teacher()
t.intro()
t.show_teacher()
s = Student()
s.intro()
s.show_teacher()
s.show_student()
```

```python
#3. Hierarchial Inheritance
class Animal:
    def speak(self):
        print("Speak method.")

class Cat(Animal):
    pass

class Dog(Animal):
    pass
```

```python
#4. Multiple Inheritance
class VC:
    def emails(self):
        print("VC Class.")

class Registrar:
    def workload(self):
        print("Registrar Class.")

class Student(VC, Registrar):
    pass

s = Student()
s.emails()
s.workload()
```

### Polymorphism

```python
#1. Function Polymorphism
print(len("Atika"))
print(len([1, 2, 3, 4, 5]))
print(len((1, 2, 3, 4)))
```

```python
#2. Operator Polymorphism
print(3 + 5)
print("Atika" + " Mughal")
```

```python
#3. Run time polymorphism (Method Overriding)
class University:
    def work(self, data):
        pass

class AI(University):
    def work(self, data):
        return "AI Department."

class MMG(University):
    def work(self, workload):
        return "MMG Department."

university = [AI(), MMG()]
a = AI()
a.work("data")
```

```python
#4. Compile polymorphism is not directly supported but achieved via *args
class Methods:
    def __init__(self, *args):
        if len(args) == 2:
            print("Two Arguments.")
        elif len(args) == 3:
            print("Three Arguments.")
        elif len(args) == 4:
            print("Four Arguments.")
        else:
            print("More than four.")
m = Methods(1, 2)
m = Methods(2, 3, 4)
m = Methods(1, 2, 3, 4, 5)
m = Methods(1, 2, 3)
```

### Abstraction

```python
from abc import ABC, abstractmethod

class University(ABC):
    @abstractmethod
    def teacher(self):
        pass

class Student(University):
    def teacher(self):
        return "Student of AI."

s = Student()
s.teacher()
```
