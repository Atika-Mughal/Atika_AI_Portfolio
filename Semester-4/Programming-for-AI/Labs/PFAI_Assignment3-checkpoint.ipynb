### Task 8

```python
marks = int(input("Enter the marks:"))
if marks >= 90:
    print("A+")
elif marks >= 80:
    print("A")
elif marks >= 70:
    print("B")
elif marks >= 60:
    print("C")
else:
    print("Fail")
```

```python
text = input("Enter any text:")
if text.isdigit():
    print("Digits only")
elif text.isalpha():
    print("Only letters")
else:
    print("Mixed")
```
