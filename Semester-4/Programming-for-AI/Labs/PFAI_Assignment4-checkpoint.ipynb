### Task 6

```python
word = input("Enter any word").lower().strip()
count = 0
for ch in word:
    if ch in "a, e, i, o, u":
        count = count +1
print("Number of vowels:", count)
```

### Task 7

```python
word = input("Enter any word")
count = 0
for ch in word:
    if ch.isdigit():
        count = count +1
print("Number of digits:", count)
```

### Task 8

```python
word = input("Enter any string:")
rev = ""
for ch in word:
    rev = ch + rev
print("Reversed String:", rev)
```

### Task 10

```python
pin = "4321"
attempts = 0
while attempts < 3:
    entry = input("Enter PIN:")
    if entry == pin:
        print("Unlocked.")
        break
    attempts += 1
else:
    print("Blocked")
```

### Task 11

```python
word = input("Enter any word:")
for i in range(len(word)):
    if word[i].lower() == 'a':
        print("First 'a' is at position:", i)
        break
else:
    print("Not found")
```

### Small Mini-Project

```python
secret_number = 7
guess_number = int(input("Enter any guess number"))
while guess_number != secret_number:
    if guess_number > secret_number:
        print("Too high")
        guess_number = int(input("Try again."))
    else:
        print("Too low")
        guess_number = int(input("Try again."))
print("Correct!")
```
