### Task 3

```python
basic_salary = int(input("Enter your salary"))
house_rent = basic_salary * 30 / 100
medical = basic_salary * 10 / 100
gross_salary = basic_salary + house_rent + medical
print("Basic Salary\t\t: ", basic_salary)
print("House Rent (30%)\t: ", house_rent)
print("Medical Allowance (10%) : ", medical)
print("Gross Salary\t\t: ", gross_salary)
```

### Task 4

```python
age = int(input("Enter your age: "))
marks = int(input("Enter your intermediate marks: "))
result = age >= 18 and marks >= 60
print("Eligible for Admission: ", result)
```

### Task 5

```python
total_marks = int(input("Enter the total marks: "))
obtained = int(input("Enter the obtained marks: "))
percentage = obtained * 100 / total_marks
print("Percentage: ", percentage)
passed = percentage >= 50
print("Passed: ", passed)
distinction = percentage >= 80
print("Distinction: ", distinction)
```

### Task 6

```python
print("-"*30)
print("\tRESULT CARD")
print("-"*30)

name = input("Enter your name: ")
roll_number = input("Enter your Roll Number: ")
print("Name\t: ", name)
print("Roll No\t: ", roll_number)

subject1 = int(input("Enter Subject1 Marks: "))
subject2 = int(input("Enter Subject2 Marks: "))
subject3 = int(input("Enter Subject3 Marks: "))

print("\nMarks: ")
print("Subject 1 : ", subject1)
print("Subject 2 : ", subject2)
print("Subject 3 : ", subject3)

total_marks = 300
obtained = subject1 + subject2 + subject3
print("\nTotal Marks: ", obtained, "/ 300")
percentage = obtained * 100 / total_marks
print("Percentage: ", percentage)

passed = percentage >= 50
print("\nPassed\t: ", passed)
distinction = percentage >= 80
print("Distinction: ", distinction)
```
