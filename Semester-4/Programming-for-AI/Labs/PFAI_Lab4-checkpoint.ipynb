### Task 1

```python
def employee_salary(basic_salary, allowances, bonus, tax_rate, insurance_rate):
    gross_salary = basic_salary + allowances + bonus
    tax_deduction = gross_salary * (tax_rate/100)
    insurance_deduction = gross_salary * (insurance_rate/100)
    net_salary = gross_salary - (tax_deduction + insurance_deduction)
    annual_salary = net_salary * 12
    return gross_salary, tax_deduction, insurance_deduction, net_salary, annual_salary
basic_salary = float(input("Enter the basic salary: "))
allowances = float(input("Enter Allowances: "))
bonus = float(input("Enter Bonus: "))
tax_rate = float(input("Enter Tax Rate % : "))
insurance_rate = float(input("Enter the Insurance Rate % : "))

gross_salary, tax_deduction, insurance_deduction, net_salary, annual_salary = employee_salary(basic_salary, allowances, bonus, tax_rate, insurance_rate)

print("\n-----Salary Slip-----")
print("Gross Salary: ", gross_salary)
print("Tax Deuction: ", tax_deduction)
print("Insurance Deduction: ", insurance_deduction)
print("Net Salary: ", net_salary)
print("Annual Salary: ", annual_salary)
```

### Task 2

```python
 # *reports = function can accept many dictionaries
def evaluate_models (* reports):
    total_accuracy = 0
    best_model = None
    worst_model = None

    highest_f1 = -1
    highest_loss = -1
    for report in reports:
        total_accuracy += report["accuracy"]
        if report["f1_score"] > highest_f1:
            highest_f1 = report["f1_score"]
            best_model = report["model_name"]

        if report["loss"] > highest_loss:
            highest_loss = report["loss"]
            worst_model = report["model_name"]
            
    avg_accuracy = total_accuracy / len(reports)
    return avg_accuracy, best_model, worst_model

model1 = {
    "model_name": "Decision Tree",
    "accuracy": 0.91,
    "precision": 0.88,
    "recall": 0.86,
    "f1_score": 0.87,
    "loss": 0.25
}
model2 = {
    "model_name": "Random Forest",
    "accuracy": 0.94,
    "precision": 0.90,
    "recall": 0.91,
    "f1_score": 0.92,
    "loss": 0.18
}
model3 = {
    "model_name": "SVM",
    "accuracy": 0.89,
    "precision": 0.85,
    "recall": 0.84,
    "f1_score": 0.85,
    "loss": 0.30
}
avg_accuracy, best_model, worst_model = evaluate_models(model1, model2, model3)
print("Average Accuracy:", avg_accuracy)
print("Best Model (F1-Score):", best_model)
print("Worst Model (Loss):", worst_model)
```
