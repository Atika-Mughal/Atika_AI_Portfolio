### Task 3

```python
temperature = int(input("Enter the temperature in celsius:"))
if temperature >= 35:
    print("Heatwave Alert!")
elif temperature >= 25:
    print("Warm Weather") 
elif temperature >= 15:
    print("Pleasant Weather")
else:
    print("Cold Weather")
```

### Task 10

```python
n_students = int(input("Enter the number of students:"))
n_days = int(input("Enter the number of days:"))
for i in range(n_students):
    present = 0
    absent = 0
    for j in range(n_days):
        attendance = input("Enter A or P:")
        if attendance == 'P':
            present += 1
        else:
            absent +=1
    print("Present:", present)
    print("Absent:", absent)
```
