```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Task-1

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
print(df.isnull().sum())
```

```python
Age = df['Age'].dropna()
```

```python
Age.isna().sum()
```

```python
plt.figure(figsize=(6,5))
plt.title("Age of Passengers")
plt.xlabel("Time")
plt.ylabel("Values")
plt.grid(True)
plt.xticks(rotation=45)
plt.hist(Age)
plt.show()
```

### Task-2

```python
n_passengers = df.groupby("Sex")["Sex"].count()
print(n_passengers)
```

```python
bar = plt.bar(n_passengers.index, n_passengers.values, color=['grey', 'lightgreen'])
plt.title("Number of Passengers")
plt.xlabel("Data")
plt.ylabel("Values")
plt.bar_label(bar)
plt.show()
```

### Task-3

```python
g_class = df.groupby("Pclass")["Pclass"].count()
print(g_class)
```

```python
avg = df.groupby("Pclass")["Fare"].mean()
print(avg)
```

```python
plt.bar(avg.index, avg.values, color=['grey', 'lightgreen', 'blue'])
plt.title("Fare of each class")
plt.xlabel("Class")
plt.ylabel("Mean")
plt.show()
```

### Task-4

```python
port = df.groupby("Embarked")["Fare"].mean()
print(port)
```

```python
plt.barh(port.index, port.values, color=['indigo', 'red', 'black'])
plt.title("Average Fare by Embarked")
plt.xlabel("Values")
plt.ylabel("Embarked Values")
plt.show()
```
